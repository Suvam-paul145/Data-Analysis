```python
fifa = pd.read_csv('fifa_data.csv')

fifa.display()
```

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.arange(0,110,10)
plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#abcdef' )
    # change hatch
    #box.set(hatch = '/')

plt.show()
```

```python
# prompt: create one graph of elipse

t = np.linspace(0, 2*np.pi, 100)
a = 5  # Semi-major axis
b = 3  # Semi-minor axis
x = a * np.cos(t)
y = b * np.sin(t)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.title('Ellipse')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')  # Ensures the ellipse is not distorted
plt.show()
```

```python
# prompt: create a graph of sin and cos of taking random value

# Generate random x values
x = np.random.rand(100) * 2 * np.pi

# Calculate sin(x) and cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Sort the values by x for a clearer plot
sorted_indices = np.argsort(x)
x_sorted = x[sorted_indices]
y_sin_sorted = y_sin[sorted_indices]
y_cos_sorted = y_cos[sorted_indices]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_sorted, y_sin_sorted, label='sin(x)', color='blue')
plt.plot(x_sorted, y_cos_sorted, label='cos(x)', color='red')

# Add title and labels
plt.title('Sine and Cosine of Random Values')
plt.xlabel('Random Value (radians)')
plt.ylabel('Function Value')

# Add a legend
plt.legend()

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: create graph of tanx

x_tan = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 200) # Avoid vertical asymptotes
y_tan = np.tan(x_tan)

plt.figure(figsize=(8, 6))
plt.plot(x_tan, y_tan, label='tan(x)', color='green')

plt.title('Graph of tan(x)')
plt.xlabel('x (radians)')
plt.ylabel('tan(x)')

plt.grid(True)
plt.ylim(-10, 10) # Limit the y-axis to avoid extreme values near asymptotes

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create subplots: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --------- Parabola: y = x^2 ---------
x = np.linspace(-5, 5, 100)
y = x**2

axes[0].plot(x, y, label='y = x', color='purple')
axes[0].set_title('Graph of a Parabola')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()
axes[0].grid(True)

# --------- tan(x) ---------
x_tan = np.linspace(-np.pi/2 + 0.1, np.pi/2 - 0.1, 200)  # Avoid asymptotes
y_tan = np.tan(x_tan)

axes[1].plot(x_tan, y_tan, label='tan(x)', color='green')
axes[1].set_title('Graph of tan(x)')
axes[1].set_xlabel('x (radians)')
axes[1].set_ylabel('tan(x)')
axes[1].grid(True)
axes[1].set_ylim(-10, 10)  # Limit y-axis
axes[1].legend()

# Show the final figure
plt.tight_layout()
plt.show()
```
