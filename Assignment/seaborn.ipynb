```python
# Part 1: Setup & Exploration (10 mins) 
# Tasks: 
# 1. Import seaborn, pandas, and matplotlib.pyplot. 
# 2. Load the built-in tips dataset. 
# 3. Display the first 5 rows using .head(). 
# 4. Check basic stats using .describe(). 

# Step 1: Import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load built-in 'tips' dataset
tips = sns.load_dataset('tips')

# Step 3: Display first 5 rows
print(tips.head())

# Step 4: Show basic statistics
print(tips.describe())
```

```python
# Part 2: Distribution Plots (15 mins) 
# Tasks: 
# 1. Create a histogram of total_bill using sns.histplot(). 
# 2. Plot a KDE (smoothed curve) for tip using sns.kdeplot(). 
# 3. Create a combined distribution plot with histogram + KDE for total_bill.

# Step 1: Histogram of total_bill
sns.histplot(tips['total_bill'], kde=False, bins=20)
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# Step 2: KDE plot of tip
sns.kdeplot(tips['tip'], shade=True)
plt.title("KDE of Tip")
plt.xlabel("Tip Amount")
plt.grid()
plt.show()

# Step 3: Combined Histogram + KDE
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title("Total Bill Distribution with KDE")
plt.xlabel("Total Bill")
plt.grid()
plt.show()
```

```python
# Part 3: Categorical Plots (20 mins) 
# Tasks: 
# 1. Use sns.boxplot() to show total_bill across different days. 
# 2. Use sns.violinplot() to compare total_bill distribution by sex. 
# 3. Create a stripplot() of total_bill by day, colored by sex, with jitter=True. 
# 4. Bonus: Combine a violinplot() and stripplot() using ax = plt.gca(). 

# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# --- Task 1: Boxplot of total_bill across different days ---
plt.figure(figsize=(6, 4))
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.grid()
plt.tight_layout()
plt.show()

# --- Task 2: Violinplot of total_bill by sex ---
plt.figure(figsize=(6, 4))
sns.violinplot(x='sex', y='total_bill', data=tips)
plt.title("Violinplot of Total Bill by Sex")
plt.xlabel("Sex")
plt.ylabel("Total Bill")
plt.grid()
plt.tight_layout()
plt.show()

# --- Task 3: Stripplot of total_bill by day, colored by sex ---
plt.figure(figsize=(6, 4))
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', jitter=True, dodge=True)
plt.title("Stripplot of Total Bill by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.legend(title='Sex')
plt.grid()
plt.tight_layout()
plt.show()

# --- Task 4 (Bonus): Combine violinplot and stripplot ---
plt.figure(figsize=(6, 4))
ax = plt.gca()
sns.violinplot(x='day', y='total_bill', data=tips, inner=None, ax=ax)
sns.stripplot(x='day', y='total_bill', data=tips, color='k', size=3, jitter=True, ax=ax)
plt.title("Combined Violinplot and Stripplot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Part 4: Relationship Plots (15 mins) 
# Tasks: 
# 1. Create a scatterplot() of total_bill vs tip, colored by smoker. 
# 2. Use sns.pairplot() with hue='sex' to explore relationships. 
# 3. Use sns.jointplot() (kind='reg') to show total_bill vs tip with regression.

# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset('tips')

# --- Task 1: Scatterplot of total_bill vs tip, colored by smoker ---
plt.figure(figsize=(6, 4))
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips)
plt.title("Scatterplot: Total Bill vs Tip by Smoker")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.grid()
plt.tight_layout()
plt.show()

# --- Task 2: Pairplot with hue='sex' ---
sns.pairplot(tips, hue='sex')
plt.suptitle("Pairplot of Tips Dataset by Sex", y=1.02)
plt.show()

# --- Task 3: Jointplot of total_bill vs tip with regression ---
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
plt.show()
```

```python
# Part 5: Correlation Heatmap (10 mins) 
# Tasks: 
# 1. Load the built-in titanic dataset. 
# 2. Drop any non-numeric columns. 
# 3. Compute the .corr() matrix. 
# 4. Plot a heatmap with: 
# o annot=True 
# o cmap="coolwarm" 
# o linewidths=0.5

# Step 1: Import required libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the built-in Titanic dataset
titanic = sns.load_dataset("titanic")

# Step 3: Drop non-numeric columns
titanic_numeric = titanic.select_dtypes(include='number')

# Step 4: Compute the correlation matrix
corr_matrix = titanic_numeric.corr()

# Step 5: Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap - Titanic Dataset")
plt.show()
```
